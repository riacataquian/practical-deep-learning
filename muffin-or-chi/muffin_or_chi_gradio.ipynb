```python
import fastbook
fastbook.setup_book()
```

```python
from pathlib import Path
from fastai.vision.all import *

path = Path()
path.ls(file_exts='.pkl')

learn_inf = load_learner(path/'muffin-or-chi.pkl')
```

```python
import gradio as gr

def predict(img):
    pred, pred_idx, probs = learn_inf.predict(img)
    return {
        learn_inf.dls.vocab[i]: float(probs[i])
        for i in range(len(probs))
    }

demo = gr.Interface(
    fn=predict,
    inputs=gr.Image(type="pil", label="Upload an image"),
    outputs=gr.Label(num_top_classes=3, label="Predictions"),
    title="Muffin or Chi?",
    description="Image classification inference using a Fast.ai-trained model"
)

if __name__ == "__main__":
    demo.launch(share=True)
```
